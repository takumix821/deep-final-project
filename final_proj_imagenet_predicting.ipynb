### Importing the necessary modules

```python
import numpy as np
import pandas as pd
import os
import glob
from os import listdir
from os.path import isfile, join
import cv2
```

### Loading and Preprocessing

```python
data_path1 = 'E://orchid_private_set'
fnames1 = glob.glob(data_path1 + '/*.jpg')

print("found",len(fnames1),"jpg files.")
```

```python
data_path2 = 'E://orchid_public_set'
fnames2 = glob.glob(data_path2 + '/*.jpg')

print("found",len(fnames2),"jpg files.")
```

```python
#Load image
x_private = []

for i in range(len(fnames1)):
    file_ = data_path1 + '/' + fnames1[i]
    img = cv2.imread(file_,cv2.IMREAD_COLOR)
    img = cv2.resize(img, (128, 128), interpolation=cv2.INTER_AREA) # 方形
    x_private.append(img)

print(np.array(x_private).shape)
```

```python
#Load image
x_public = []

for i in range(len(fnames2)):
    file_ = data_path2 + '/' + fnames2[i]
    img = cv2.imread(file_,cv2.IMREAD_COLOR)
    img = cv2.resize(img, (128, 128), interpolation=cv2.INTER_AREA) # 方形
    x_public.append(img)

print(np.array(x_public).shape)
```

```python
from tensorflow.keras.models import load_model
model_final_2 = load_model("D://my_model63.hdf5")
```

```python
model_final_2.summary()
```

```python
y_private = model_final_2.predict(x_private)
```

```python
# 將output改為標籤
output_private = np.zeros(y_private.shape[0])

for i in range(y_private.shape[0]):
    for j in range(y_private.shape[1]):
        if y_private[i, j] >= max(y_private[i, ]):
            output_private[i] = j

print(output_private[:5])
print(len(output_private))
```

```python
y_public = model_final_2.predict(x_public)
```

```python
# 將output改為標籤
output_public = np.zeros(y_public.shape[0])

for i in range(y_public.shape[0]):
    for j in range(y_public.shape[1]):
        if y_public[i, j] >= max(y_public[i, ]):
            output_public[i] = j

print(output_public[:5])
print(len(output_public))
```

```python
df1 = pd.dataframe({'filename': fnames1, 'category': output_private})
df2 = pd.dataframe({'filename': fnames2, 'category': output_public})

df = pd.concat([df1, df2])

df.to_csv('D:/OneDrive/110_02/deep/final proj/predict.csv', index=False)
```
